# Phase 3: Feature Engineering
## Create Derived Features for Enhanced Analysis
**Objective:** Engineer meaningful features from existing data
**Features:** Rating categories, cost categories, cuisine analysis, service flags

```python
# Import Libraries
import pandas as pd
import numpy as np

# Load cleaned dataset from Phase 2
df_engineered = pd.read_csv('Zomato_datasets/zomato_cleaned.csv')

print("✓ Cleaned data loaded from Phase 2. Starting feature engineering...")
print(f"Dataset shape: {df_engineered.shape}")
```

```python
# Feature 1: Rating Categories
def categorize_rating(rate):
    if pd.isna(rate):
        return 'Unknown'
    elif rate >= 4.5:
        return 'Excellent'
    elif rate >= 4.0:
        return 'Very Good'
    elif rate >= 3.5:
        return 'Good'
    elif rate >= 3.0:
        return 'Average'
    else:
        return 'Below Average'

df_engineered['rating_category'] = df_engineered['rating'].apply(categorize_rating)
print(" Feature 1: Rating categories created")
print(df_engineered['rating_category'].value_counts())
```

```python
# Feature 2: Cost Categories (Quantile-based binning)
df_engineered['cost_category'] = pd.qcut(df_engineered['avg_cost'], 
                                          q=4, 
                                          labels=['Budget', 'Mid-Range', 'Premium', 'Luxury'],
                                          duplicates='drop')
print("\n Feature 2: Cost categories created (Budget/Mid-Range/Premium/Luxury)")
print(df_engineered['cost_category'].value_counts())
```

```python
# Feature 3: Cuisine Count and Primary Cuisine
df_engineered['cuisine_count'] = df_engineered['cuisines'].apply(
    lambda x: len(str(x).split(',')) if pd.notna(x) else 0
)

df_engineered['primary_cuisine'] = df_engineered['cuisines'].apply(
    lambda x: str(x).split(',')[0].strip() if pd.notna(x) else 'Not Specified'
)

print("\n Feature 3: Cuisine count and primary cuisine extracted")
print(f"Cuisine count range: {df_engineered['cuisine_count'].min()} - {df_engineered['cuisine_count'].max()}")
print(f"Top 10 Primary Cuisines:")
print(df_engineered['primary_cuisine'].value_counts().head(10))
```

```python
# Summary of Engineered Features
print("\n")
print("ENGINEERED FEATURES SUMMARY:"+"")

print(f"\nNew Features Created:")
print(f"  1. rating_category - Categorical ratings (Excellent/Very Good/Good/Average/Below Average)")
print(f"  2. cost_category - Cost tiers (Budget/Mid-Range/Premium/Luxury)")
print(f"  3. cuisine_count - Number of cuisines offered")
print(f"  4. primary_cuisine - Main cuisine type")
print(f"\nTotal engineered features: {len(df_engineered.columns)}")
print(f"\nDataset shape after feature engineering: {df_engineered.shape}")
print("\n Feature engineering complete!")
```

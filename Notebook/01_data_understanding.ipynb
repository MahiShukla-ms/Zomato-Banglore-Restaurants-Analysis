# Phase 1: Data Understanding & Exploration
## Zomato Restaurant Dataset Analysis
**Objective:** Load, explore, and understand the structure of the Zomato dataset
**Dataset:** 7105 restaurants with 12 columns

```python
# Import Required Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
# Configure display options
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)
plt.rcParams['figure.figsize'] = (12, 6)
print(" Libraries imported successfully !")
```

```python
# Load the dataset
df = pd.read_csv('Zomato_datasets/zomato.csv', encoding='latin-1')
print(" Dataset loaded successfully !")
print(f"\nDataset Shape: {df.shape}")
print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")
```

```python
# Display column names and types
print("COLUMN NAMES & DATA TYPES:"+"\n")
print(df.dtypes)
print(f"Total Columns: {len(df.columns)}")
```

```python
# Display first few rows
print("FIRST 10 ROWS:"+"\n")
print(df.head(10).to_string())
```

```python
# Detailed dataset info
print("DATASET INFO:")
print("="*70)
df.info()
```

```python
# Missing values analysis
print("MISSING VALUES ANALYSIS:")
print("="*70)
missing_data = pd.DataFrame({
    'Column': df.columns,
    'Missing_Count': df.isnull().sum(),
    'Missing_Percent': (df.isnull().sum() / len(df) * 100).round(2)
})
missing_data = missing_data.sort_values('Missing_Percent', ascending=False)
print(missing_data.to_string())
print(f"\nTotal Missing Values: {df.isnull().sum().sum()}")
```

```python
# Basic statistics
print("BASIC STATISTICS:")
print("="*70)
print(df.describe().to_string())
```

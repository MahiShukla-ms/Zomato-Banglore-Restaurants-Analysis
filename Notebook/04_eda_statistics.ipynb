# Phase 4: Exploratory Data Analysis (EDA) & Statistics
## Univariate, Bivariate, and Multivariate Statistical Analysis
**Objective:** Understand data distributions and relationships
**Methods:** Descriptive stats, correlations, cross-tabulations

```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load cleaned dataset from Phase 2
df = pd.read_csv('Zomato_datasets/zomato_cleaned.csv')

# Add engineered features (if not already present)
if 'rating_category' not in df.columns:
    df['rating_category'] = df['rating'].apply(
        lambda x: 'Excellent' if x >= 4.5 else ('Very Good' if x >= 4.0 else ('Good' if x >= 3.5 else ('Average' if x >= 3.0 else 'Below Average')))
    )
if 'cost_category' not in df.columns:
    df['cost_category'] = pd.qcut(df['avg_cost'], q=4, labels=['Budget', 'Mid-Range', 'Premium', 'Luxury'], duplicates='drop')
if 'cuisine_count' not in df.columns:
    df['cuisine_count'] = df['cuisines'].apply(lambda x: len(str(x).split(',')) if pd.notna(x) else 0)

print("✓ Cleaned data loaded and prepared for statistical analysis")
print(f"Dataset shape: {df.shape}")
```

```python
# 3. Cross-Tabulation Analysis
print("\n")
print("CROSS-TABULATION: Restaurant Type vs Rating Category"+"\n")

crosstab_type_rating = pd.crosstab(df['restaurant_type'], df['rating_category'], margins=True)
print(crosstab_type_rating)

print("\n" + "="*70)
print("CROSS-TABULATION: Rating Category vs Online Order Availability")
print("="*70)
crosstab_rating_online = pd.crosstab(df['rating_category'], df['has_online_order'], margins=True)
print(crosstab_rating_online)
```

```python
# 2. Correlation Analysis
print("\n" + "="*30)
print("CORRELATION ANALYSIS")
print("="*30)
numeric_cols = ['rating', 'num_ratings', 'avg_cost', 'has_online_order', 'has_table_booking', 'cuisine_count']
corr_matrix = df[numeric_cols].corr()
print(corr_matrix)
```

```python
# 4. Groupby Statistics
print("\n" + "="*30)
print("STATISTICS BY RATING CATEGORY")
print("="*30)
rating_stats = df.groupby('rating_category').agg({
    'rating': ['mean', 'median', 'std', 'count'],
    'num_ratings': 'mean',
    'avg_cost': 'mean',
    'has_online_order': 'mean',
    'has_table_booking': 'mean'
}).round(2)
print(rating_stats)

print("\n")
print(" Statistical analysis complete!")
```
